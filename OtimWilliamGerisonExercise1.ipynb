```python
#Importing Libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
#Defining our Data
#Relationship is y = 3x + 1
#Cleaned data by removing 5 since it did not correspond to any y value
x = np.array([-1, 0, 1, 2, 3, 4], dtype=int)
y = np.array([-2, 1, 4, 7, 10, 13], dtype=int)
```

```python
#Defining the model
neuralmodel = keras.Sequential([keras.layers.Dense(units=1, input_shape = [1])])
```

```python
#Compiling Our Models
from typing import Optional
neuralmodel.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#Training the Model
neuralmodel.fit(x,y, epochs=1000)
```

```python
print(neuralmodel.predict([500]))
# When x= 34, y = 102.99991 ~ 103
# When x = 103, y = 309.9997 ~ 310
# When x = 500, y = 1500.9985 ~ 1501
```
